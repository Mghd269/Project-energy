# Analysis of Renewable Energy Production in Poland

## Introduction

Electricity is indispensable for the functioning of modern society. In Poland, as well as in many other countries worldwide, electricity production relies heavily on fossil fuels such as coal (both hard and brown coal), natural gas, and oil. However, fossil fuels are a significant source of air pollution and greenhouse gas emissions, contributing to climate change.

In response to these challenges, the development of renewable energy has gained prominence. Renewable energy is derived from natural sources such as sunlight, wind, water, and biomass. It is a clean energy alternative that does not produce pollutants or greenhouse gases.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
RES = pd.read_excel('source_of_energy.xlsx', 'RenewableEnergySources')
CES = pd.read_excel('source_of_energy.xlsx', 'ConventionalEnergySources')
```

```python
RES = pd.DataFrame(RES)
CES = pd.DataFrame(CES)
RES = RES.drop(['2001','2002', '2003', '2004'], axis=1)
CES = CES.drop(['2001','2002', '2003', '2004'], axis=1)
```

```python
RES.replace(to_replace='-', value=0, inplace=True)
poland_data_RES = RES[RES['Voivodeship'] == "Poland"]
years = poland_data_RES.columns[1:]
values = poland_data_RES.iloc[:, 1:].values.flatten()
```

```python
plt.plot(years, values.astype(float), marker='o', linestyle='-', color='green')
plt.title('Data for RES Poland in years 2005-2022')
plt.xlabel('Year')
plt.ylabel('Value in MWh')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
```

## Renewable Energy Progress in Poland

Poland has made substantial strides in harnessing renewable energy sources (RES) in recent years.
Key initiatives and programs have played a pivotal role in fostering the development of RES.

### Background (2001-2004)

During the period from 2001 to 2004, 
Poland was in the early stages of exploring and conceptualizing renewable energy.
At that time, there were virtually no consolidated renewable energy sources in the Polish energy system.
The country heavily relied on conventional sources, including coal, gas, and oil.

### Transition and Initiatives

Over the years, there has been a noticeable shift in focus,
driven by increasing environmental awareness and societal pressure for sustainable development.
In response, Poland introduced various support mechanisms to encourage investment in renewable energy projects.
These measures included green certificate schemes and feed-in tariffs.

### Current Scenario

Presently, Poland is experiencing a significant increase in the role of renewable sources in its energy mix.
This includes wind, solar, biomass, and geothermal energy.
The transition towards more sustainable energy sources is crucial both environmentally and economically.

### Future Considerations

Continued efforts are essential to develop infrastructure and technologies related to renewable energy sources.
This will further increase their share in the energy mix,
contributing to a more environmentally friendly and sustainable energy system.

As you can see, Poland's energy sector relies on CES. The reason for this is that in Poland, there is a significant emphasis on the development and utilization of energy from various sources. This commitment to a mix of energy resources contributes to the resilience and sustainability of Poland's energy infrastructure, ensuring a stable and reliable supply for the nation's needs.

## Energy Landscape in Poland: Progress in Renewable Energy

In Poland, there exists a significant disproportion between conventional energy sources (CES) and renewable energy sources (RES).
Historically, CES has dominated the energy landscape, overshadowing the contribution of RES.

However, a positive trend is evident in the data, 
indicating progress in the adoption of renewable energy over the years.
The increasing utilization of RES suggests a concerted effort to diversify 
the energy mix and reduce reliance on conventional sources.

This shift aligns with global sustainability goals and reflects Poland's commitment to a more environmentally friendly and resilient energy sector.
Despite the current disproportion, the emphasis on RES signals a promising trajectory towards a more balanced and sustainable energy future for Poland.

# Renewable Energy Leadership in West Pomeranian Voivodeship

## Introduction

West Pomeranian Voivodeship stands out as the most renewable energy-powered region in Poland, a distinction attributed to its abundant wind and solar resources, coupled with unwavering commitment to renewable energy development. This study delves into the region's prominent position in renewable energy generation and its implications for Poland's transition to a cleaner energy future.

## Key Insights

* **Wind Energy Dominance:** West Pomeranian Voivodeship harnesses wind power as the primary renewable energy source. Wind farms across the region generate substantial amounts of electricity, solidifying its position as a wind energy powerhouse.

* **Solar Energy Prominence:** Solar energy also plays a significant role in the region's renewable energy portfolio. A growing number of solar panels installed on homes, businesses, and public buildings contributes to the region's solar energy production.

* **Biomass as a Sustainable Fuel:** Biomass power plants utilize wood and other organic materials to generate electricity, further diversifying the region's renewable energy mix.

# Countries grouped by GDP
he bar chart above, in my opinion, is challenging to interpret and draw conclusions from. Therefore, I intend to enhance its clarity by sorting the data into groups based on their GDP. This approach aims to facilitate a more straightforward comparison of the percentage of renewable energy production across these groups.

```python
gdp = pd.read_excel("nama_10_gdp_page_spreadsheet.xlsx", "Sheet 1")
gdpdf = pd.DataFrame(gdp)
```

```python
gdpdf.drop(gdpdf.columns[2:19:2], axis=1, inplace=True)
gdpdf.drop(index=[0],inplace=True)
gdpdf.rename(columns={"TIME": "Countries"}, inplace=True)
gdpdf
gdpdf.replace(":", None, inplace=True)
gdpdf.fillna(0, inplace=True)
```

```python
gdp_2021 = gdpdf["2021"]
dfeu["Gdp2021"] = gdp_2021.values
```

```python
dfeu["2021"].astype(float)
dfeu.sort_values(by="Gdp2021", ascending=True, inplace=True)
```

```python
Q1 = dfeu['Gdp2021'].quantile(0.25)
Q2 = dfeu['Gdp2021'].quantile(0.5)
Q3 = dfeu['Gdp2021'].quantile(0.75)
```

```python
quantiles = [float('-inf'), Q1, Q2, Q3, float('inf')]
labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']
dfeu["Group"] = pd.cut(dfeu["Gdp2021"], bins=quantiles, labels=labels)
dfeu.reset_index(drop=True)
```

```python
group_1 = dfeu.loc[dfeu["Group"] == "Group 1", :]
group_2 = dfeu.loc[dfeu["Group"] == "Group 2", :]
group_3 = dfeu.loc[dfeu["Group"] == "Group 3", :]
group_4 = dfeu.loc[dfeu["Group"] == "Group 4", :]
colors2 = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', 
           '#7f7f7f', '#bcbd22']
```

```python
Res_2021_eu =  group_4["2021"]
countries = group_4["Countries"]
plt.bar(countries, Res_2021_eu, color=colors2)
plt.xticks(rotation=45, fontsize=7, ha="right")
plt.title("RES in Europe")
plt.xlabel('Country')
plt.ylabel('Percent of total energy production')
plt.legend()
plt.grid(True)
```
